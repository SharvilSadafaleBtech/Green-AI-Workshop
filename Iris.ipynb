```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


data = {
"Solar": [1200, 1500, 1300],
"Wind": [3400, 3600, 3200],
"Hydropower": [2900, 3100, 2800],
"Blomass": [2500, 2700, 2400]
}

df =pd.DataFrame(data)

sns.pairplot(df)
plt.show()
```

```python
#create a heatmap to show the correlation matrix
sns.heatmap(df.corr(), annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.boxplot(data)
plt.title("Distribution of Energy Consumption By Source")
plt.show()
```

```python
sns.violinplot(data=df)
plt.show()
```

```python
ev = [100,200,300,400,500]
ce = [10,20,25,40,50]

dfr = pd.DataFrame({'Energy Consumption' : ev,
'Carbon Emissions' : ce})
sns.regplot(x = 'Energy Consumption',y = 'Carbon Emissions', data = dfr)
plt.show()
```

```python
from mpl_toolkits import mplot3d
fig=plt.figure()
ax=plt.axes(projection='3d')
z = np.linspace(0,1,1000)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
ax.plot3D(x,y,z,'green')
ax.set_title('3D line plot')
plt.show()
```

```python
fp = "/content/IRIS.csv"
df = pd.read_csv(fp)
df.head()
sns.pairplot(df, hue='species', palette='bright')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()

```

```python
correlation_matrix = df.iloc[:, :-1].corr() # Exclude 'species' column for correlation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Feature Correlation')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Boxplot of Sepal Length by Species')
plt.show()
```

```python

sns.violinplot(x='species', y='petal_length', data=df)
plt.title('Violinplot of Petal Length by Species')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
df['species_encoded'] = lab.fit_transform(df['species'])

mat = df.drop(columns=['species']).corr()
sns.heatmap(mat, annot=True, cmap='coolwarm',linewidths=0.5)
plt.show()
```
